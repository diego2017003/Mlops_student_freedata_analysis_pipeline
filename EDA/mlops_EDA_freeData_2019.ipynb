```python
import pandas as pd
import numpy as np
from plotly.subplots import make_subplots
import plotly.graph_objects as go
import plotly.express as px
```

```python
dados_2019 = pd.read_csv('/content/drive/MyDrive/Dados mlops/dados_merge_discentes_20192.csv',sep=';',encoding='utf-8')
```

```python
dados_2019.info()
```

```python
d = dados_2019['sexo'].str.replace('false','OUTROS').value_counts()
fig = px.bar(d)
fig.update_layout(yaxis=dict(title=None),xaxis=dict(title="Gênero"))
fig.show()
```

```python
d = dados_2019['raca'].value_counts()
fig = px.bar(d,x=d.index,y=d.values,color_discrete_sequence=['#fff555','#efefdd'])
fig.update_layout(yaxis=dict(title=None,gridcolor="#ff0000"),xaxis=dict(title="Raça"),plot_bgcolor='#ffffff',annotations=True)
fig.show()
```
